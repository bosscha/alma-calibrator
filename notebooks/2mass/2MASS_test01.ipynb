```python
Irsa.list_catalogs()
```

```python
result = Irsa.query_region("3C 279", catalog="fp_psc", spatial="Cone", radius= 12 * u.arcmin)

result
```

```python
ra_2mass = result['ra'].data.data # "mask-type" -_-
dec_2mass = result['dec'].data.data
```
