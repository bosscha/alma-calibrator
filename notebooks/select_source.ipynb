- we can make a selection using selectDeepfield for data

- or using selectDeepfield_fromsql for accessing saved sql data

### Selection process

Select sources with criteria:

- ignore freq res (for imaging)

- ignore polarization product

- excludeCycle0 data

- only accept data from 12m array (or 12m7m)

- minimum integration time per band is 1h for B3, B6, and B7 (after filtering all above)

```python
report = q.selectDeepfield_fromsql("calibrators_gt_0.5Jy.db", maxFreqRes=10000000, array='12m', \
        excludeCycle0=True, selectPol=False, minTimeBand={3:3600., 6:3600., 7:3600.}, verbose=True, silent=True)
```

#### write the report in a file

```python
q.writeReport(report, "report.txt", silent=True)
```
