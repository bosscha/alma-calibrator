```python
import sys
sys.path.append('../src/')

from ALMAQueryCal import *
```

### Select calibrator with Flux > 0.5 Jy

```python
fileCal = "callist_20170329.list"
```

```python
q       = queryCal(fileCal, fluxrange = [0.5, 1000.0])
```

```python
print "Number of objects: ", len(q.listCal)
```

```python
q.listCal
```

### Query the observation/project from ALMA archive

Some will fail

```python
data    = q.queryAlma(q.listCal, public = True)
```

### Select calibrator which has polarization XY, ignore the integration time and freq res

- set 0 for minTimeBand 
- set a large number in maxFreqRes

```python
report  = q.selectDeepField(data, minTimeBand = {3:0.0, 6:0.0, 7:0.0}, maxFreqRes = 10000000.0, selectPol=True, verbose = True)
q.writeReport(report, file = fileCal+".pol.report")
```

### Select calibrator with total integration time > 5h 
Ignore the maxFreqRes (for imaging)

```python
report  = q.selectDeepField(data, minTimeBand = {3:18000.0, 6:18000.0, 7:18000.0}, maxFreqRes = 10000000.0, verbose = True)
q.writeReport(report, file = fileCal+".longIntegration.report")
```

#### Integration > 5h with highRes

```python
report  = q.selectDeepField(data, minTimeBand = {3:18000.0, 6:18000.0, 7:18000.0}, maxFreqRes = 1000.0, verbose = True)
q.writeReport(report, file = fileCal+".longIntegration_abs.report")
```

### Rather small integration time + highres (for absorption lines)


```python
report  = q.selectDeepField(data, minTimeBand = {3:500.0, 6:500.0, 7:3600.0}, maxFreqRes = 1000.0, verbose = True)
q.writeReport(report, file = fileCal+".large_abs.report")
```

#### ignoring maxFreqRes

```python
report  = q.selectDeepField(data, minTimeBand = {3:500.0, 6:500.0, 7:3600.0}, maxFreqRes = 10000000.0, verbose = True)
q.writeReport(report, file = fileCal+".large.report")
```
